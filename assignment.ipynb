```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index to include Date as a column
tesla_data.reset_index(inplace=True)

# Display the first 5 rows
tesla_data.head()
```

```python
# Plot Tesla's closing price over time
plt.figure(figsize=(10, 6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla Close Price')
plt.title("Tesla Closing Stock Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd

# Load the Tesla revenue data
revenue_df = pd.read_csv("tesla_revenue.csv")

# Display the first few rows
print(revenue_df.head())
```

```python
# Rename columns for clarity
revenue_df.columns = ["Year", "Revenue"]

# Remove the dollar sign and commas from the 'Revenue' column and convert to numeric
revenue_df["Revenue"] = revenue_df["Revenue"].replace({"[$,]": ""}, regex=True).astype(float)

# Display the cleaned data
print(revenue_df.head())
```

```python
import matplotlib.pyplot as plt

# Plot Tesla's annual revenue
plt.figure(figsize=(10, 6))
plt.bar(revenue_df["Year"], revenue_df["Revenue"], color="skyblue")
plt.title("Tesla Annual Revenue Over the Years")
plt.xlabel("Year")
plt.ylabel("Revenue (Millions of US $)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
revenue_df.to_csv("cleaned_tesla_revenue.csv", index=False)
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download the Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2022-12-31')

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data.index, data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.grid(True)
    plt.show()

# Call the make_graph function
make_graph(tesla_data, 'Tesla Stock Price')
```
